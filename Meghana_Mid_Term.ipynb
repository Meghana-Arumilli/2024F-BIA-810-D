## Step 0. Prepare raw input datasets

Here we will 1) load original datasets, 2) remove unused columns, 3) de-duplicate rows, and 4) join datasets, not necessarily in this order

Assumptions: 
- Claim ID (cur_clm_uniq_id) represents one claim, which may or may not have more than one code (code could be HCPCS/CPT, diagnosis, procedure...)
- There is a one-to-many relationship between patient IDs (bene_mbi_id) and claim IDs (cur_clm_uniq_id), i.e. each claim is unique to one patient, but one patient can have more than one claim

### 0.3 Combine all datasets

### 0.3.1. Join datasets with common records

```python
# Join Claims Header and Claims Revenue Center datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    parta_claims_header_df,
    parta_claims_revenue_center_detail_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

# Classifying HCPs into Segments based on the claim count

```python
def classify_hcp(row):
    if row['claim_count'] == 1:
        return 'Disease Aware'
    elif 2 <= row['claim_count'] <= 4:
        return 'Trialists'
    elif 5 <= row['claim_count'] <= 9:
        return 'Rising Stars'
    elif row['claim_count'] >= 10:
        return 'High-Volume Prescribers'
    else:
        return 'Other'

hcp_claim_counts['segment'] = hcp_claim_counts.apply(classify_hcp, axis=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
claims_df = pd.read_csv('parta_claims_header.csv')

# Step 2: Define cardiovascular-related diagnosis codes (example ICD-10 codes)
cvm_diagnosis_codes = ['I10', 'I20', 'I25', 'I50']  # Replace or expand with relevant codes

# Step 3: Add a new column for claim year extracted from 'clm_from_dt' (assuming it's a date column)
claims_df['claim_year'] = pd.to_datetime(claims_df['clm_from_dt']).dt.year

# Step 4: Filter the dataset to keep only CVM claims using 'prncpl_dgns_cd' or 'admtg_dgns_cd'
cvm_claims_df = claims_df[
    claims_df['prncpl_dgns_cd'].str.startswith(tuple(cvm_diagnosis_codes)) |
    claims_df['admtg_dgns_cd'].str.startswith(tuple(cvm_diagnosis_codes))
]

# Step 5: Group by 'region' (using 'prvdr_fac_fips_st_cd' as proxy) and 'claim_year' to count CVM claims per region
cvm_claims_per_region_year = cvm_claims_df.groupby(['prvdr_fac_fips_st_cd', 'claim_year']).size().reset_index(name='cvm_claim_count')

# Step 6: Calculate total CVM claims per year across all regions
total_cvm_claims_per_year = cvm_claims_per_region_year.groupby('claim_year')['cvm_claim_count'].sum().reset_index(name='total_cvm_claims')

# Step 7: Merge dataframes to calculate market share for each region
cvm_claims_merged_df = pd.merge(cvm_claims_per_region_year, total_cvm_claims_per_year, on='claim_year')
cvm_claims_merged_df['cvm_market_share'] = (cvm_claims_merged_df['cvm_claim_count'] / cvm_claims_merged_df['total_cvm_claims']) * 100

# Step 8: Filter to keep only regions with the highest average market share over the period
top_regions = (
    cvm_claims_merged_df.groupby('prvdr_fac_fips_st_cd')['cvm_market_share']
    .mean()
    .sort_values(ascending=False)
    .head(5)  # Top 5 regions with the highest average market share
    .index
)

filtered_cvm_claims_merged_df = cvm_claims_merged_df[cvm_claims_merged_df['prvdr_fac_fips_st_cd'].isin(top_regions)]

# Step 9: Pivot the data for visualization
cvm_claims_pivot = filtered_cvm_claims_merged_df.pivot(index='claim_year', columns='prvdr_fac_fips_st_cd', values='cvm_market_share')

# Step 10: Plotting the market share trends for the top regions
plt.figure(figsize=(12, 8))
for region in cvm_claims_pivot.columns:
    plt.plot(cvm_claims_pivot.index, cvm_claims_pivot[region], label=region)

# Step 11: Add labels, title, and legend to the plot
plt.xlabel('Year')
plt.ylabel('CVM Market Share (%)')
plt.title('Top Regions CVM Claims Market Share by Year (2016-2018)')
plt.xticks([2016, 2017, 2018])  # Set x-axis ticks to show only the years
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Patient Risk Segmentation and Resource Allocation

## How do the risk levels of patients (e.g., presence of co-morbidities) affect the volume and types of CVM claims?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load claims dataset
claims_df = pd.read_csv('parta_claims_header.csv')  # Claims dataset

# Step 1: Define cardiovascular and co-morbidity related diagnosis codes
# Example ICD-10 codes for co-morbidities (extend this list as needed)
cvm_diagnosis_codes = ['I10', 'I20', 'I25', 'I50']  # Cardiovascular diagnoses
diabetes_codes = ['E08', 'E09', 'E10', 'E11', 'E13']  # Example codes for diabetes
hypertension_codes = ['I10', 'I11', 'I12', 'I13', 'I15']  # Hypertension
copd_codes = ['J44']  # COPD example code

# Step 2: Add columns to indicate presence of co-morbidities
claims_df['has_diabetes'] = claims_df['prncpl_dgns_cd'].str.startswith(tuple(diabetes_codes)) | claims_df['admtg_dgns_cd'].str.startswith(tuple(diabetes_codes))
claims_df['has_hypertension'] = claims_df['prncpl_dgns_cd'].str.startswith(tuple(hypertension_codes)) | claims_df['admtg_dgns_cd'].str.startswith(tuple(hypertension_codes))
claims_df['has_copd'] = claims_df['prncpl_dgns_cd'].str.startswith(tuple(copd_codes)) | claims_df['admtg_dgns_cd'].str.startswith(tuple(copd_codes))

# Step 3: Create a co-morbidity count column to calculate the number of co-morbidities for each claim
claims_df['comorbidity_count'] = claims_df[['has_diabetes', 'has_hypertension', 'has_copd']].sum(axis=1)

# Step 4: Categorize patients into risk levels based on the number of co-morbidities
def categorize_risk_level(row):
    if row['comorbidity_count'] == 0:
        return 'Low Risk'
    elif row['comorbidity_count'] <= 2:
        return 'Medium Risk'
    else:
        return 'High Risk'

claims_df['risk_level'] = claims_df.apply(categorize_risk_level, axis=1)

# Step 5: Filter CVM claims and group by risk level
cvm_claims_df = claims_df[
    claims_df['prncpl_dgns_cd'].str.startswith(tuple(cvm_diagnosis_codes)) |
    claims_df['admtg_dgns_cd'].str.startswith(tuple(cvm_diagnosis_codes))
]

# Group by risk level to see the count of CVM claims
risk_level_claims = cvm_claims_df.groupby('risk_level').size().reset_index(name='cvm_claim_count')

# Step 6: Plotting the impact of risk level on CVM claims
plt.figure(figsize=(10, 6))
plt.bar(risk_level_claims['risk_level'], risk_level_claims['cvm_claim_count'], color=['green', 'orange', 'red'])
plt.xlabel('Risk Level')
plt.ylabel('Number of CVM Claims')
plt.title('CVM Claim Volume by Patient Risk Level')
plt.show()

# Step 7: Further Analysis - Types of Claims by Risk Level
# Assuming we have a column 'claim_type' that indicates the type of CVM claim (e.g., 'Procedure', 'Hospital Admission')
if 'claim_type' in cvm_claims_df.columns:
    claim_types_by_risk = cvm_claims_df.groupby(['risk_level', 'claim_type']).size().unstack().fillna(0)

    # Plotting the types of claims by risk level
    claim_types_by_risk.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='viridis')
    plt.xlabel('Risk Level')
    plt.ylabel('Number of CVM Claims by Type')
    plt.title('Types of CVM Claims by Patient Risk Level')
    plt.legend(title='Claim Type')
    plt.tight_layout()
    plt.show()
```
